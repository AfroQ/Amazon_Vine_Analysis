```python
import pandas as pd
```

```python
# Read in vine_table as a dataframe
vine_df = pd.read_csv("vine_table.csv", index_col=0)
vine_df
```

```python
filtered_vine_df = vine_df[vine_df["total_votes"] >= 20]
filtered_vine_df
```

```python
helpful_percentage_df = filtered_vine_df[(filtered_vine_df["helpful_votes"]/filtered_vine_df["total_votes"])*100 >=50]
helpful_percentage_df
```

```python
paid_reviews_df = helpful_percentage_df[helpful_percentage_df["vine"] == "Y"]
paid_reviews_df.head(10)
```

```python
unpaid_reviews_df = helpful_percentage_df[helpful_percentage_df["vine"] == "N"]
unpaid_reviews_df.head(10)
```

```python
# Total reviews
total_reviews = len(helpful_percentage_df)
# Total five star reviews
five_star_reviews = len(helpful_percentage_df[helpful_percentage_df["star_rating"] == 5.0])
# Total paid reviews
total_paid = len(paid_reviews_df)
# Total unpaid reciews
total_unpaid = len(unpaid_reviews_df)
# Total five star reviews - paid
total_five_star_paid = len(paid_reviews_df[paid_reviews_df["star_rating"] == 5.0])
# Total five star reciews - unpaid
total_five_star_unpaid = len(unpaid_reviews_df[unpaid_reviews_df["star_rating"] == 5.0])

# Percent total paid five star reviews
five_star_paid_percentage = f"{total_five_star_paid/total_paid * 100:.0f}%"
five_star_unpaid_percentage = f"{total_five_star_unpaid/total_unpaid * 100:.0f}%"
five_star_unpaid_percentage
```
